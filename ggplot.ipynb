* Create a function to convert ºF to ºC and in (inches) to mm.

```r
# Write code here
FtoC <- function(v){ return (v - 32)/18000}
inToMm <- function(v){ return (v * 25.4)}

FtoC(data[2])
inToMm(data[9])
```
